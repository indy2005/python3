# Strings

## Basic string operations

```python
# strings are iterable
name = "Stephen"
print(name[0])
```

```python
for c in name:
    print(c)
```

```python
# strings are immutable
try:
    name[1] = "S"
except TypeError as e:
    print("Error: ", e)
```

### Boolean string methods


```python
name = "Stephen"
name.isupper()
```

```python
name = "Stephen"
name.islower()
```

```python
name = "Stephen"
print(name.istitle())
name = "Stephen mackenzie"
print(name.istitle())
```

```python
name = " "
name.isspace()
```

```python
price = "99.5"
print(price.isnumeric())
price = "99.5f"
print(price.isnumeric())
```

```python
sentence = "I once stood in a field. It was 23 years ago." # full stops and spaces = False
print(sentence.isalnum())
sentence = "IoncestoodinafieldItwas23yearsago" # no full stops or spaces = True
print(sentence.isalnum())
sentence = "Ioncestoodinafield!Itwas23yearsago" # puncutation mark (!) = False
print(sentence.isalnum())
```

```python
sentence = "I once stood in a field. It was 23 years ago."  # numbers, whitespace and punctuation = False
print(sentence.isalpha())
sentence = "IoncestoodinafieldItwas23yearsago" # numbers = False
print(sentence.isalpha())
sentence = "IoncestoodinafieldItwastwentythreeyearsago" # no numbers, punctuation or whitepsace = True
print(sentence.isalpha())
sentence = "Ioncestoodinafield Itwastwentythreeyearsago" # whitepsace = False
print(sentence.isalpha())
```

## Joining Strings

```python
# basic string concatenation with +
first_name = "Stephen"
last_name = "Mackenzie"
full_name = first_name + " " + last_name
full_name
```

```python
# using join
names = ",".join(["Fred", "Mary", "Stephen", "Jack"])
names
```

## Splitting Strings

```python
name = "Fred Mary Jane Stephen Jack"
names = name.split(" ")
names
```

## Slicing strings

```python
name = "Stephen Mackenzie"
print(name[0]) # first character
print(name[0:5]) # first 5 characters (up to index 4 inclusive, 5 non-inclusive)
print(name[:]) # print the entire string
print(name[-3:]) # last 3 characters
print(name[::-1]) # reverse a string
```

## Stripping strings

```python
# strip whitepsace from the start and end of a string
name = " Stephen Mackenzie "
name.strip()
```

## Searching within a string

```python
# find the first occurence of a string inside another string
sentence = "Once upon a time in Hollywood, yes I did say Hollywood..."
sentence.find('Hollywood')
```

```python
# if not found, returns a -1
sentence = "Once upon a time in Hollywood, yes I did say Hollywood..."
sentence.find('Elephant')
```

```python
# ignoring case
# find the first occurence of a string inside another string
sentence = "Once upon a time in Hollywood, yes I did say Hollywood..."
find_me = "HollyWood"
sentence.lower().find(find_me.lower())
```

## Replacing parts of a string

```python
# replace all occurences of one item with another
sentence = "Once upon a time in Hollywood, yes I did say Hollywood..."
replace_me = "Hollywood"
replace_with = "Manchester"
sentence.replace(replace_me, replace_with)
```
